## Define Functions

```python
def MCC(omega_true, robsel_omega):
    
    #Get off-diagonal entries
    p = robsel_omega.shape[0]
    omega_true = omega_true[np.tril_indices(p,-1)]
    omega_hat = omega_hat[np.tril_indices(p,-1)]
    
    #MCC
    TP = np.sum((omega_true!=0.) & (omega_hat!=0.))
    FN = np.sum((omega_true!=0.) & (omega_hat==0.))
    FP = np.sum((omega_true==0.) & (omega_hat!=0.))
    TN = np.sum((omega_true==0.) & (omega_hat==0.))
    MCC = (TP*TN - FP*FN)/np.sqrt((TP+FP)*(TP+FN)*(TN+FP)*(TN+FN))
    return MCC

def method_comparision(robsel_omega, true_omega):
    p = robsel_omega.shape[0]
    #RobSel
    robsel_edges = robsel_omega[np.tril_indices(p,-1)]
    true_omega = true_omega[np.tril_indices(p,-1)]
    robsel_fp = np.sum((robsel_edges!=0) & (true_omega==0))
    robsel_tp = np.sum((robsel_edges!=0) & (true_omega!=0))
    robsel_tn = np.sum((robsel_edges==0) & (true_omega==0))
    robsel_fn = np.sum((robsel_edges==0) & (true_omega!=0))
        
    jaccard_table = np.array([robsel_fp, robsel_tp, robsel_tn, robsel_fn])
    return jaccard_table
```

## Robust Selection

```python
lambd = RobustSelection(df, alpha=0.05, B=200, with_diag=True)
lambd
```

```python
robsel_model = GraphicalLasso(alpha=lambd)
robsel_model.fit(df)
Robsel_Omega = robsel_model.precision_
```
